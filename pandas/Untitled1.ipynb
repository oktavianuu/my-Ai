```python
import pandas as pd
reviews = pd.read_csv("data/winemag-data-130k-v2.csv", index_col=0)
reviews.transpose()
```

```python
# we can access dataframe property by using dot (.), here we access country in our dataframe
reviews.country
```

```python
# other than that, we can also use square brackets to access it
reviews['country']
```

```python
# The square brackets method has advantages over the dot method since we can pass multiple values in it
# we can access the first vaalue in country by doing so
reviews['country'][0]
```

```python
# INDEXING IN PANDAS
# FOR more advance indexing, we use loc and iloc
# two paradigms when indexing in pandas
# # iloc select data based on its numerical position in the data
reviews.iloc[0]
```
