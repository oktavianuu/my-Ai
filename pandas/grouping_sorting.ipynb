```python
import pandas as pd
reviews = pd.read_csv("/home/oktavianu/data/winemag-data-130k-v2.csv", index_col=0)
```

```python
# We can replicate the value_counts() by doing the following:
reviews.groupby('points').points.count()
```

```python
# We can think of each group we generate as being a slice of the DataFrame containing only data with values that match. The value is accessible 
# directly to us using apply() method, and we can then manipulate it.

reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# or even more fine-grained control, you can also group by more than one column. For an example, here's how we would pick out the best wine
# by country and province:
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# Another groupby() method worth mentioning is agg(), which lets you run a bunch of different functions on your DataFrame simultaneously. 
# For example, we can generate a simple statistical summary of the dataset as follows:
reviews.groupby("country").price.agg([len, min, max])
```
