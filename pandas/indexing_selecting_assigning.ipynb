```python
import pandas as pd
reviews = pd.read_csv("/home/oktavianu/data/winemag-data-130k-v2.csv", index_col=0)
reviews.transpose()
```

```python
# we can access dataframe property by using dot (.), here we access country in our dataframe
reviews.country
```

```python
# other than that, we can also use square brackets to access it
reviews['country']
```

```python
# The square brackets method has advantages over the dot method since we can pass multiple values in it
# we can access the first vaalue in country by doing so
reviews['country'][0]
```

```python
# INDEXING IN PANDAS
# FOR more advance indexing, we use loc and iloc
# two paradigms when indexing in pandas
# # iloc select data based on its numerical position in the data
reviews.iloc[0]
```

```python
# iloc works based on row, so when we call iloc, it will retrieve data based on row.
# if we want to retriece column, we can do so:
reviews.iloc[:, 0]
```

```python
# the ":" means we retrieved everything in the selected column
# if we want to select certain range of entries, let say 1, 2, 3 entris in country column, we do the following:
reviews.iloc[:4, 0]
```

```python
# or select the second and the third entries
reviews.iloc[1:3, 0]
```

```python
# also possible to pass a list
reviews.iloc[[0, 1, 2, 3], 0]
```

```python
# negative argument can be used to retrieve three rows from last, last --> fwd
reviews.iloc[-3:]
```

```python
# LABEL BASED SELECTION; loc
# we used the label, not numerical position
reviews.loc[0, 'country'] # retrieve the first row in column "country"
```

```python
reviews.loc[:, ['country', 'taster_name', 'taster_twitter_handle', 'points']]
```

```python
# When to use loc or iloc?
# iloc follows python rule --> 0:10 will select 0, ..., 9, excluding 10. 
# loc meanwhile can index any type including string, if we want to retrieve all column lets say between columns "Apple" and "Potatoe"
# we can do this: loc["Apples":"Potatoes"]
```

```python
# Manipulating index
# We can use any column as index using set_index()
reviews.set_index('title')
```

```python
# conditional selection means that if we need to retrieve data based on certain conditions, we can do the following
# For example, suppose that we're interested specifically in better-than-average wines produced in Italy.
# We can start by checking if each wine is Italian or not:
reviews.country=="Italy"
```

```python
# The result above produced a series of boolean values based on country. The result now can be used inside of loc to select relevan data
reviews.loc[reviews.country=="Italy"]
```

```python
# now we see that there are 19540 rows based on country "Italy"
```

```python
# We can also select more advance data, lets say for example we want to select wine made in Italy with more than 90 points
reviews.loc[(reviews.country=="Italy") & (reviews.points >= 90)]
```

```python
# Suppose we'll buy any wine that's made in Italy or which is rated above average. For this we use a pipe (|):
reviews.loc[(reviews.country=="Italy") | (reviews.points >= 90)]
```

```python
# we can use isin to select data whose values "is in" a list of values.
# Here for example we select wines only from Italy and France
reviews.loc[reviews.country.isin(["Italy", "France"])]
```

```python
# notnull and isnull, you know the function from its name
reviews.loc[reviews.price.notnull()]
```
